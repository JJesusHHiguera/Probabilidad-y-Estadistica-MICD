1) Alice tiene 2 hijos y uno de ellos es niña. ¿Cuál es la probabilidad de que el otro hijo también sea niña?

```python
# Creamos una lista con las posibilidades que solo pueden ser dos: hijo o hija
posib = ['hijo', 'hija']

# Representemos el conjunto universo de todas las combinaciones posibles
hijos = {(i,j) for i in posib for j in posib}

# La siguiente es la lista donde se guardarán los casos favorables
fav = []

for m, n in hijos:
    if m == 'hija':
        if n == m:
            fav.append((n,m))

'''
 Calculamos la probabilidad
 NOTA: a partir de este punto, las variables para representar probabilidad, los casos favorables y
 el conjunto universo, van a estar seguidas del número de problema al que están asociados
'''

prob_1 = len(fav)/len(hijos)

print("La probabilidad de que el otro hijo también sea niña es de:", prob_1)
```

2) Un dado justo se tira 2 veces. ¿Cuál es la probabilidad de obtener un 2 en la primera tirada y no tener un 4 en la segunda tirada? 

```python
# Seleccionamos un rango de 1 a 6 para representar las caras del dado
dado = range(1,7)

#Definimos el conjunto universo
univ_2 = {(i,j) for i in dado for j in dado}

# Hagamos un proceso similar al anterior para el caso expuesto en este problema
casos_2 = []

for tirada_1, tirada_2 in univ_2:
    if tirada_1 == 2:
        if tirada_2 != 4:
            casos_2.append((tirada_1, tirada_2))
        
prob_2 = len(casos_2)/len(univ_2)

print(f"La probabilidad es de {prob_2:.4f}")
```

8) Se tiene la moneda A y la moneda B. La moneda A tiene 1/2 de probabilidad de salir águila o sol en cada tirada. La probabilidad de sacar sol en la moneda B es 1/3. Si seleccionamos una moneda aleatoriamente y la tiramos hasta obtener sol. La probabilidad de seleccionar la moneda A es 1/4 y la probabilidad de seleccionar la moneda B es 3/4. ¿Cuál es el número esperado de veces que tiraremos la moneda para obtener sol?

```python
# Hagamos el procedimiento por pasos. Primero encontremos el valor esperado para la moneda A
# La probabilidad de obtener sol en la moneda A es:
prob_sA = 1/2

# El número esperado de lanzamientos es
VE_A = 1/prob_sA

#La probabilidad de escoger esta moneda es de:
prob_A = 1/4

# Análogamente, para la moneda B:
prob_sB = 1/3
VE_B = 1/prob_sB
prob_B = 3/4

# Ahora, procedemos a calcular la cantidad total de lanzamientos esperados. Seguiremos la siguiente fórmula:
# E(total) = P(A)E_A + P(B)E_B
# donde P(A) y P(B) son las probabilidades de escoger las monedas A y B, respectivamente
VE_total = prob_A*VE_A + prob_B*VE_B

print(f"El número esperado de lanzamientos es de {VE_total}")
```

9) Una compañía de seguros vende un seguro de vida que paga 240,000 pesos y la póliza es válida por 1 año. Le vende el seguro a una joven de 25 años por 210 pesos. La probabilidad de que la cliente sobreviva el año es de 0.999592. ¿Cuál es el valor esperado de esta póliza para la compañía de seguros?

```python
# Comencemos por definir las variables asociadas a las cantidades dadas por el problema
seg = 240000 # La cantidad que paga el seguro
pol = 210 # El precio de la póliza
P_v = 0.999592 # La probabilidad de sobrevivir
P_m = 1 - P_v # Probabilidad de fallecer

# Ahora, vamos a determinar la cantidad que pagaría el seguro en caso de que la joven fallezca
# Notemos que los cálculos están hechos desde el punto de vista de la compañía, por lo tanto, el precio de la póliza es un ingreso y el del seguro
# es una pérdida
total = pol - seg

# Podemos calcular el valor esperado multiplicando cada resultado por la probabilidad de ese resultado para luego sumar los productos
val_esp = (total*P_m)+(pol*P_v)

print(f"El valor esperado de esta póliza para la compañía de seguros es una ganancia de {val_esp:.2f} pesos")
```

13) ¿Cuál es la probabilidad de que dos estudiantes tengan su cumpleaños en el mismo día en una clase de 30 estudiantes?

```python
"""
Empecemos planteando el problema opuesto; que nadie cumpla años el mismo día. Para que esto ocurra, comencemos con una persona, la probabilidad de 
que en la clase haya alguien que no cumpla años el mismo día que él es del 100%, pues no hay nadie, algo que también se puede representar con
365/365. Ahora, la siguiente persona no puede cumplir años el mismo día y, como los eventos son independientes, la probabilidad estaría dada como
(365/365)(364/365). Si agregamos una más se tendría (365/365)(364/365)(363/365). Entonces, para n personas se tiene (365/365)(364/365)...((365-n+1)/365), 
o bien, en términos de factoriales: 365!/[(365^n)(365-n)!]. Representemos esto a continuación:
"""
# Importamos la librería math para poder ocupar el factorial
import math

# Definimos el número de personas en la clase
n = 30

prob_13 = math.factorial(365)/((365**n)*math.factorial(365-n))

# Ahora bien, para calcular la probabilidad de que al menos dos estudiantes cumplan años el mismo día, simplemente hacemos la resta correspondiente
print(f'La probabilidad de que al menos dos estudiantes cumplan años el mismo día es de {1-prob_13:.4f}')
```

14) Si estas vendiendo sándwiches y el 70% de las personas eligen de huevo y el resto de pollo. ¿Cuál es la probabilidad de vender 2 sándwiches de huevo dentro de los 3 siguientes clientes? 

```python
# Probabilidad de vender un sandwich de huevo
h = 0.7
# Probabilidad de vender uno de pollo
p = 0.3

# Sólo hay dos tipos de sándwich
sand = {'h', 'p'}

# Determinemos el espacio muestral
univ_14 = {(i,j,k) for i in sand for j in sand for k in sand}

casos_14 = []

for a in univ_14:
    if a.count('h') == 2:
        casos_14.append(a)

# Los eventos son independientes, por lo tanto, la probabilidad de que ocurran es la multiplicación de la probabilidad de cada evento
# Poesteriormente, elevamos cada probabilidad al número de sándwiches vendidos (2 de huevo y uno de pollo)
prob_14 = (h**(casos_14[0].count('h')))*p**((casos_14[0].count('p')))*len(casos_14)

print(f"La probabilidad de vender dos sándwiches de huevo dentro de los siguientes 3 clientes es de {prob_14:.3f}")
```

15) Un hospital ha registrado que el 75% de pacientes que está sufriendo una enfermedad muere debido a ella. ¿Cuál es la probabilidad de que 4 de 6 pacientes seleccionados al azar se recuperen?

```python
# Probabilidad de fallecer por la enfermedad
M = 0.75

# Probabilidad de recuperarse 
V = 0.25

# Los pacientes sólo pueden estar en dos estados
pac = {'M', 'V'}

# El conjunto universo es el conjunto de todos los casos posibles para 6 pacientes tomados de forma aleatoria
univ_15 = {(a,b,c,d,e,f) for a in pac for b in pac for c in pac
           for d in pac for e in pac for f in pac}


casos_15 = []

# Elegimos únicamente aquellos casos donde hay 4 pacientes vivos
for r in univ_15:
    if r.count('V') == 4:
        casos_15.append(r)

# Los sucesos son idependientes, pues que un paciente se recupere no influye en que el siguiente lo haga o no, entonces, la probabilidad
# estará dada por la probabilidad de fallecer elevada a la potencia del número de pacientes fallecidos (2 en este caso) por la probabilidad
# de recuperarse elevada a la potencia del número correspondiente (4, por como se define en el problema)
prob_15 = (M**casos_15[0].count('M'))*(V**casos_15[0].count('V'))

print(f"La probabilidad de que 4 de 6 pacientes seleccionados al azar se recuperen de la enfermedad es de {prob_15:.4f}")
```

18) La probabilidad de tener cierta enfermedad en una población es de 0.00148. La prueba de laboratorio para dicha enfermedad tiene una tasa de 93% de verdaderos positivos (sale positiva cuando realmente se tiene la enfermedad el 93 % de las veces). Y su tasa de verdaderos negativos es de 99 % (Sale negativa cuando no se tiene la enfermedad el 99 % de las veces). ¿Cuál es la probabilidad de tener la enfermedad dado que la prueba salió positiva? Nota: Recordar el teorema de Bayes P(A | B)P(B) = P(B|A)P(A)

```python
# Comencemos definiendo las variables que representen los valores dados por el problema
P_E = 0.00148   # Probabilidad de estar enfermo
Ver_pos = 0.93  # Tasa de verdaderos positivos
Ver_neg = 0.99  # Tasa de verdaderos negativos

# En consecuencia:
Fal_pos = 1 - Ver_pos   # Tasa de falsos positivos
Fal_neg = 1 - Ver_neg   # Tasa de falsos negativos
P_S = 1 - P_E           # Probabilidad de estar sano o de no tener la enfermedad

# De acuerdo con el teorema de Bayes, la probabilidad de tener la enfermedad dado que la prueba salió positiva estaría dada de la forma:
# P(E|+) = P(+|E)P(E)/P(+), donde P(+) es la probabilidad de que la prueba salga positiva y P(E) la probabilidad de estar enfermo.

# Sin embargo, la prueba puede salir positiva si se está o no se está enfermo, de modo que P(+) es la suma de ambas posibilidades:
# P(+) = P(S)P(+|S) + P(E)P(+|E), donde P(S) es la probabilidad de estar sano. Así:

prob_18 = (Ver_pos*P_E)/((P_S*Fal_neg)+(P_E*Ver_pos))

print(f"La probabilidad de tener la enfermedad dado que la prueba salió positiva es del {100*prob_18:.2f}%")
```

19) Haz un código que escriba letras y el espacio con la misma frecuencia con que aparecen en el cuento de Alicia en el País de las Maravillas : https://github.com/andergd/separadorSilabas/blob/master/Carroll%2C%20Lewis%20-%20Alicia%20En%20El%20Pa%C3%ADs%20De%20Las%20Maravillas.txt

```python
# Importamos la librería necesaria para leer el archivo de texto
import requests

# Damos el url tipo raw para leer adecuadamente el archivo
url = 'https://raw.githubusercontent.com/andergd/separadorSilabas/master/Carroll%2C%20Lewis%20-%20Alicia%20En%20El%20Pa%C3%ADs%20De%20Las%20Maravillas.txt'

response = requests.get(url)

# Verificamos que todo esté en orden
if response.status_code == 200:
    contenido = response.text
    print(contenido) # Visualizaremos el archivo si todo está bien
else:
    print(f"Error al obtener el archivo: {response.status_code}") # Veremos este mensaje de error si algo ha fallado
```

```python
# Ahora vamos a la siguiente parte de este ejercicio: determinar las frecuencias y escribir un texto nuevo
from collections import Counter # Usamos esta herramienta para contar cada caracter,
import string                   # esta otra para contar cada letra
import random                   # y esta otra para generar el texto de forma aleatoria

# Filtramos aquellos caracteres que nos interesa contabilizar
# Decidí añadir los saltos de línea para que, como se dijo en clase, el código pueda escribir un texto que, a simple vista parezca tener sentido. 
# De lo contrario todo saldría en una misma línea y no se podría apreciar este efecto
letras = [carac for carac in contenido if carac in string.ascii_letters + ' ' + '\n']

# Aplicamos Counter para asociar cada caracter con el número de veces que aparece en el texto
frec_letra = Counter(letras)

# Definimos la longitud del archivo
long = len(contenido)

freq = []

# Añadimos cada caracter el mismo número de veces como lo marca su respectiva frecuencia relativa
for letra, frecuencia in frec_letra.items():
    freq.extend([letra]*frecuencia)

# Podemos hacer un texto con la misma longitud "long" o con un tamaño diferente el cual podemos fijar desde un principio o pedirselo al usuario:
# tam = int(input("¿Cuántos caracteres quieres que tenga tu texto?"))
# Luego, en el siguiente comando reemplazaríamos 'long' por 'tam'

# Generamos el nuevo texto escrito a partir de las frecuencias determinadas antes
nuevo_texto = ''.join(random.choice(freq) for _ in range(long))

# Escribimos el texto correspondiente
print(nuevo_texto)
```
